```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
```

```python
from sklearn.datasets import load_iris
```

```python
#iris 데이터 셋 로드
iris_dataset = load_iris()
```

```python
#데이터셋의 키 값들 확인
print(f"iris_dataset의 키: {iris_dataset.keys()}")
```

```python
from sklearn.model_selection import train_test_split

#데이터를 훈련과 테스트로 분할, train_test_split를 이용해서 데이터가 섞이게 하여 데이터가 편향적으로 나뉘어지는 것을 방지
#random_state로 같은 시드로 주어 실험때마다 다른 값이 되는 것을 방지
#test_size로 분할 비율을 다르게 함, 기본은 75:25
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'],test_size=0.33, random_state=42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(iris_dataset['data'], iris_dataset['target'],test_size=0.2, random_state=42)
```

```python
print("X_train 크기: {}".format(X_train.shape))
print("y_train 크기: {}".format(y_train.shape))

print("X_train2 크기: {}".format(X_train2.shape))
print("y_train2 크기: {}".format(y_train2.shape))
```

```python
print("X_test 크기: {}".format(X_test.shape))
print("y_test 크기: {}".format(y_test.shape))

print("X_test2 크기: {}".format(X_test2.shape))
print("y_test2 크기: {}".format(y_test2.shape))
```

```python
#X_train 데이터를 사용해서 데이터프레임을 만듭니다.ㅏ
#열의 이름은 iris_dataset.feature_names에 있는 문자열을 사용합니다.

iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
iris_dataframe2 = pd.DataFrame(X_train2, columns=iris_dataset.feature_names)

#데이터프레임ㅇ르 사용해 y_train에 따라 색으로 구분된 산점도 행렬을 만듭니다.
#s: marker 크기
#cmap: colormap
#alpha: amout of transparency applied, 얼마나 투명하게 보일 것인가 
# 대각선은 각 특성 분포의 히스토그램


pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15), marker='o', hist_kwds = {'bins':20}, s=60, alpha=8, cmap=mglearn.cm3)

pd.plotting.scatter_matrix(iris_dataframe2, c=y_train2, figsize=(15,15), marker='o', hist_kwds = {'bins':20}, s=60, alpha=8, cmap=mglearn.cm3)


```

```python
#knn 분류기 불러오기
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1) #이웃의 수 k = 1
knn.fit(X_train, y_train) # 훈련데이터와 레이블로 모델 훈련

print("테스트 세트의 정확도: {:.2f}".format(knn.score(X_test, y_test)))

knn2 = KNeighborsClassifier(n_neighbors=1)#이웃의 수 k = 1
knn2.fit(X_train2, y_train2) # 훈련데이터와 레이블로 모델 훈련
print("테스트 세트의 정확도: {:.2f}".format(knn.score(X_test2, y_test2)))
```

```python
#knn 분류기 불러오기
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2) #이웃의 수 k = 2
knn.fit(X_train, y_train) # 훈련데이터와 레이블로 모델 훈련

print("테스트 세트의 정확도: {:.2f}".format(knn.score(X_test, y_test)))

knn2 = KNeighborsClassifier(n_neighbors=2)#이웃의 수 k = 2
knn2.fit(X_train2, y_train2) # 훈련데이터와 레이블로 모델 훈련
print("테스트 세트의 정확도: {:.2f}".format(knn.score(X_test2, y_test2)))
```

```python
#knn 분류기 불러오기
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3) #이웃의 수 k = 3
knn.fit(X_train, y_train) # 훈련데이터와 레이블로 모델 훈련

print("테스트 세트의 정확도: {:.2f}".format(knn.score(X_test, y_test)))

knn2 = KNeighborsClassifier(n_neighbors=3)#이웃의 수 k = 3
knn2.fit(X_train2, y_train2) # 훈련데이터와 레이블로 모델 훈련
print("테스트 세트의 정확도: {:.2f}".format(knn.score(X_test2, y_test2)))
```
